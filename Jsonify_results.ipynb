```python
import os
import json
import pandas as pd

cwd = os.getcwd()
chinesefilepath = cwd + '/results/final_test_result/Chinese_test_final.csv'
frenchfilepath = cwd + '/results/final_test_result/french_test_final.csv'
indianfilepath = cwd + '/results/final_test_result/Indian_test_final.csv'
japanesefilepath = cwd + '/results/final_test_result/Japanese_test_final.csv'

outputfilepath = cwd + "/results/final_test_result/final_json.json"
```

```python
df = pd.read_csv(chinesefilepath)
df.head()
```

```python
grouped = df.groupby(['id'])
```

```python
chinese_list = []
for name, group in grouped:

    select_df = grouped.get_group(name)

    reviews_json = []
    for index, row in select_df.iterrows():
        review_scores = {
            "text" :row['review'],
            "score" :float(row['svm_score'])
        }
        reviews_json.append(review_scores)
    
    resturant_dict = {
        "name" : select_df['id'].iloc[0],
        "nb_reviews" : int(select_df.size),
        "reviews" : reviews_json
    }

    chinese_list.append(resturant_dict)
```
